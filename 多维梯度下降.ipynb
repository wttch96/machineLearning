```python
import numpy as np
import matplotlib.pyplot as plt
from util import load_house_data, zscore_normalize_features


X_train, Y_train = load_house_data() # type: np.ndarray, np.ndarray
X_features = ['size(sqft)','bedrooms','floors','age']
```

```python
_, axes = plt.subplots(1, 4, figsize=(12, 3), sharey=True) # type: plt.Figure, [plt.Subplot]

for i in range(len(axes)):
    ax = axes[i] # type: plt.Subplot
    ax.scatter(X_train[:, i], Y_train)
    ax.set_xlabel(X_features[i])

axes[0].set_ylabel("Price (1000's)")
plt.show()
```
