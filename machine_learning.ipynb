# 梯度下降

$$
w = w - \alpha \frac{1}{m} \sum_{i=1}^{m} ( f_{w,b}(x^{(i)}) - y^{(i)}) x^{(i)}
\\ b = b - \alpha \frac{1}{m} \sum_{i=1}^{m} ( f_{w,b}(x^{(i)}) - y^{(i)})
$$

```python
import numpy as np

w = np.array([1.0, 2.5, -3.3])
b = 4
x = np.array([10, 20, 30])

f = np.dot(w, x) + b
f
```

```python
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])

# 初始化 b, w
b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741, -53.36032453, -26.42131618])

from compute import compute_multiple_gradient
compute_multiple_gradient(X_train, y_train, w_init, b_init)
```

```python
from compute import compute_multiple_gradient_descent, compute_multiple_cost

init_w = np.zeros_like(w_init)
init_b = 0.
iterations = 1000
alpha = 5e-7

w_f, b_f, j_history = compute_multiple_gradient_descent(
    X_train, y_train, init_w, init_b,
    compute_multiple_cost, compute_multiple_gradient,
    alpha, iterations)

w_f[-1]
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, constrained_layout=True, figsize=(16, 8))
ax1.plot(j_history)
ax3.plot(w_f)
ax2.plot(100 + np.arange(len(j_history[100:])), j_history[100:])
ax1.set_title("Cost vs. iteration");  ax2.set_title("Cost vs. iteration (tail)")
ax1.set_ylabel('Cost')             ;  ax2.set_ylabel('Cost')
ax1.set_xlabel('iteration step')   ;  ax2.set_xlabel('iteration step')
plt.show()
```

使用给定的房价 data 进行梯度下降模拟

```python
from util import load_data

data = np.array(load_data())

X_train = data[:, 0:2]
x1 = X_train[:, 0]
x2 = X_train[:, 1]
X_train[:, 0] = (x1 - x1.mean()) / (x1.max() - x1.min())
X_train[:, 1] = (x2 - x2.mean()) / (x2.max() - x2.min())
y_train = data[:, 2]
# 初始化 b, w
b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741])


init_w = np.zeros_like(w_init)
init_b = 0.
iterations = 1000
alpha = 5e-7

w_f, b_f, j_history = compute_multiple_gradient_descent(
    X_train, y_train, init_w, init_b,
    compute_multiple_cost, compute_multiple_gradient,
    alpha, iterations)


import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, constrained_layout=True, figsize=(16, 8))
ax1.plot(j_history)
ax3.plot(w_f)
ax2.plot(100 + np.arange(len(j_history[100:])), j_history[100:])
ax1.set_title("Cost vs. iteration");  ax2.set_title("Cost vs. iteration (tail)")
ax1.set_ylabel('Cost')             ;  ax2.set_ylabel('Cost')
ax1.set_xlabel('iteration step')   ;  ax2.set_xlabel('iteration step')
plt.show()
```
